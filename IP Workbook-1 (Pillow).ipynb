```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#load an image
def load_image(file_path):
    image = Image.open(file_path)
    return image

#display an image
def display_image(image):
    plt.imshow(image)
    plt.axis('off') # hide the axes
    plt.show()
    
#save an image in different formats
def save_image (image, output_path, format):
    image.save (output_path, format=format)
    
#main function to demonstrate the process
def main():
    
#file path
    input_file_path= r"C:\Users\aakan\OneDrive\Desktop\AIDTM\trim 4\1.jpg"
    output_file_path_jpeg= r"C:\Users\aakan\OneDrive\Desktop\AIDTM\trim 4\1.jpeg"
    output_file_path_png= r"C:\Users\aakan\OneDrive\Desktop\AIDTM\trim 4\1.png"
    output_file_path_tiff=r"C:\Users\aakan\OneDrive\Desktop\AIDTM\trim 4\1.tiff"

#load image
    image=load_image(input_file_path)

#display the image
    display_image(image)
    
#save image
    save_image(image,output_file_path_jpeg,'JPEG')
    save_image(image,output_file_path_png,'PNG')
    save_image(image,output_file_path_tiff,'TIFF')
    
#calling the main function directly
main()
```
