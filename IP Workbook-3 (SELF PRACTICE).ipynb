```python
from PIL import Image

def image_to_monochrome(image_path):
  """Converts an image to monochrome using Pillow.

  Args:
    image_path: Path to the image file.

  Returns:
    The monochrome image as a PIL Image object.
  """

  with Image.open(image_path) as img:
    # Convert to grayscale
    gray_img = img.convert('L')

    # Optionally apply thresholding for black and white
    # thresholded_img = gray_img.point(lambda p: p > 127 and 255)

    return gray_img

# Example usage:
image_path = r"C:\Users\aakan\OneDrive\Desktop\AIDTM\trim 4\IMAGE PROCESSING\1.jpg"
monochrome_image = image_to_monochrome(image_path)

# Save the monochrome image
monochrome_image.save("monochrome_image.jpg")

print(monochrome_image)
```

```python
import cv2
import numpy as np

def image_to_monochrome(image_path):
  """Converts an image to monochrome.

  Args:
    image_path: Path to the image file.

  Returns:
    The monochrome image as a NumPy array.
  """

  # Load the image
  img = cv2.imread(image_path)

  # Convert to grayscale
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  # Apply thresholding (optional) for black and white
  _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

  return thresh

# Example usage:
image_path = r"C:\Users\aakan\OneDrive\Desktop\AIDTM\trim 4\IMAGE PROCESSING\1.jpg"
monochrome_image = image_to_monochrome(image_path)

# Display or save the monochrome image
cv2.imshow("Monochrome Image", monochrome_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

print(monochrome_image)
```
